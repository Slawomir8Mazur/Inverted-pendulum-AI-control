```python
from database_def import Record
import matplotlib.pyplot as plt
```

```python
record_base = Record()
record_base.position_set(120)
```

```python
r1 = Record(record_base)
r2 = Record(record_base)
r3 = Record(record_base)
r4 = Record(record_base)
```

```python
force_time = (0, 20)
```

```python
r1.move([force_time], dt_min=0.005)
r2.move([force_time], dt_min=0.01)
r3.move([force_time], dt_min=0.02)
r4.move([force_time], dt_min=0.05)
```

```python
plt.figure(figsize=(20, 15))
plt.plot(r1.last_movement['t'], r1.last_movement['fi'], 'r')
plt.plot(r2.last_movement['t'], r2.last_movement['fi'], 'g')
plt.plot(r3.last_movement['t'], r3.last_movement['fi'], 'b')
plt.plot(r4.last_movement['t'], r4.last_movement['fi'], 'y')
plt.legend(['dt=0.005', 'dt=0.01', 'dt=0.02', 'dt=0.05'])
plt.xlabel('time[ms]')
plt.ylabel('angle[rad]')
plt.title('accuracy comparison')
plt.savefig('rysunek.png')
plt.show()
```
