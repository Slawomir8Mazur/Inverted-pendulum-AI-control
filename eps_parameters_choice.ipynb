```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
descende_factor1 = 0.999
descende_factor2 = 0.998
eps = 0.9
N = np.arange(1000)
plt.figure(figsize=(12, 10))
plt.plot(N, eps*np.power(descende_factor1, N))
plt.plot(N, eps*np.power(descende_factor2, N))
plt.xlabel('epoch')
plt.ylabel('eps')
plt.legend(['descende_factor = 0.999', 'descende_factor = 0.999'])
plt.show()
```
